```python
# 겹치는 컬럼 삭제
train = train.drop(['직장시도명','입회일자_신용','최종카드발급일자','대표청구서수령지구분코드','연체일자_B0M'],axis=1)
test = test.drop(['직장시도명','입회일자_신용','최종카드발급일자','대표청구서수령지구분코드','연체일자_B0M'],axis=1)
```

```python
# 수치형 데이터만 뽑기
numeric_train_data = train.select_dtypes(include=[np.number])
numeric_test_data = test.select_dtypes(include=[np.number])
```

```python
# 범주형 데이터만 뽑기
categorical_train_data = train.select_dtypes(include='object')
categorical_test_data = test.select_dtypes(include='object')
```

```python
# 수치형 컬럼 결측치 처리 ( 중앙값 대체 )
numeric_train_data.columns[numeric_train_data.isnull().any()].tolist()
numeric_test_data.columns[numeric_train_data.isnull().any()].tolist()
```

```python
train['혜택수혜율_R3M']
```

```python
train['혜택수혜율_R3M'] = train['혜택수혜율_R3M'].fillna(train['혜택수혜율_R3M'].median())
test['혜택수혜율_R3M'] = test['혜택수혜율_R3M'].fillna(test['혜택수혜율_R3M'].median())
```

```python
# 범주형 데이터 결측치 처리 ( 최빈값 대체 )
categorical_train_data.columns[categorical_train_data.isnull().any()].tolist()
categorical_test_data.columns[categorical_test_data.isnull().any()].tolist()
```

```python
# train 데이터프레임의 결측값을 최빈값으로 대체
for column in train.columns:
    if train[column].isnull().any():
        train[column] = train[column].fillna(train[column].mode()[0])

# test 데이터프레임의 결측값을 최빈값으로 대체
for column in test.columns:
    if test[column].isnull().any():
        test[column] = test[column].fillna(test[column].mode()[0])
```

```python
# 다시 입력시키기
categorical_train_data = train.select_dtypes(include='object')
categorical_test_data = test.select_dtypes(include='object')

numeric_train_data = train.select_dtypes(include=[np.number])
numeric_test_data = test.select_dtypes(include=[np.number])
```

```python
# Robust Scale 적용
from sklearn.preprocessing import RobustScaler

Rb = RobustScaler()
Rb.fit(numeric_train_data)
Rb.transform(numeric_train_data)
Rb.transform(numeric_test_data)
```

```python
# 라벨인코딩
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# train 데이터프레임의 범주형 데이터에 대해 라벨 인코딩
for column in categorical_train_data.columns:
    categorical_train_data[column] = label_encoder.fit_transform(categorical_train_data[column])

# test 데이터프레임의 범주형 데이터에 대해 라벨 인코딩
for column in categorical_test_data.columns:
    categorical_test_data[column] = label_encoder.fit_transform(categorical_test_data[column])
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score
import matplotlib.pyplot as plt

# 1) 피처 합치기
X_train = pd.concat([numeric_train_data, categorical_train_data], axis=1)
X_test  = pd.concat([numeric_test_data,  categorical_test_data ], axis=1)

# 2) 타깃 지정 (실제 컬럼명으로 변경할 것)
y_train = train['Segment']

# 3) 학습/검증 셋 분리
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

# 4) LightGBM 데이터셋 생성
dtrain = lgb.Dataset(X_tr, label=y_tr)
dvalid = lgb.Dataset(X_val, label=y_val, reference=dtrain)

# 5) 하이퍼파라미터 설정
params = {
    'objective': 'binary',              # 이진분류
    'metric': ['auc', 'binary_logloss'],
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'max_depth': -1,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'seed': 42,
}

# 6) 모델 학습
gbm = lgb.train(
    params,
    dtrain,
    num_boost_round=1000,
    valid_sets=[dtrain, dvalid],
    valid_names=['train','valid'],
    early_stopping_rounds=50,
    verbose_eval=100
)

# 7) 테스트 데이터 예측
y_pred_proba = gbm.predict(X_test, num_iteration=gbm.best_iteration)
y_pred = (y_pred_proba > 0.5).astype(int)  # 임계치 0.5 기준 레이블 변환

# 8) 검증 셋 성능 확인
y_val_proba = gbm.predict(X_val, num_iteration=gbm.best_iteration)
print(f"Validation AUC: {roc_auc_score(y_val, y_val_proba):.4f}")
print(f"Validation Accuracy: {accuracy_score(y_val, (y_val_proba>0.5).astype(int)):.4f}")

# 9) Feature Importance 시각화
lgb.plot_importance(gbm, max_num_features=20, importance_type='gain')
plt.title("Feature Importance (Gain)")
plt.tight_layout()
plt.show()
```
