```python
# int 타입 컬럼 선택 (int16, int32, int64 포함)
int_columns = df.select_dtypes(include=['int64','float64']).columns

def compute_iqr_bounds(series):
    """IQR 방법으로 하한과 상한 계산"""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

def compute_mad_bounds(series, threshold=3.5):
    """
    MAD 방법을 이용해 하한과 상한을 계산.
    Modified Z-Score에서 임계값(threshold)을 적용하는 경우,
    |0.6745*(x - median)/MAD| <= threshold를 만족하는 범위로 변환.
    """
    median = series.median()
    mad = np.median(np.abs(series - median))
    if mad == 0:
        # MAD 값이 0이면 데이터가 상수이거나 거의 동일하므로 경계를 median으로 처리
        return median, median
    # 역산: |x - median| <= (threshold * MAD / 0.6745)
    delta = threshold * mad / 0.6745
    lower_bound = median - delta
    upper_bound = median + delta
    return lower_bound, upper_bound

# 결과 저장을 위한 딕셔너리 (컬럼별 이상치 정보)
outlier_results = {}

for col in int_columns:
    series = df[col]
    # IQR 방식으로 경계 산출
    lb_iqr, ub_iqr = compute_iqr_bounds(series)
    
    # IQR 방식의 경계가 같으면, 분포가 매우 치우치거나 상수 데이터라고 판단하여 MAD 방식 적용
    if lb_iqr == ub_iqr:
        lb, ub = compute_mad_bounds(series)
        method = "MAD"
        outliers = series[(series < lb) | (series > ub)]
    else:
        lb, ub = lb_iqr, ub_iqr
        method = "IQR"
        outliers = series[(series < lb) | (series > ub)]
        
    ratio = len(outliers) / len(series)
    
    outlier_results[col] = {
        'method': method,
        'lower_bound': lb,
        'upper_bound': ub,
        'outliers': outliers.tolist(),
        'outlier_ratio': ratio
    }
    
    print(f"컬럼: {col} - 사용 방법: {method}")
    print(f"  하한: {lb}")
    print(f"  상한: {ub}")
    print(f"  이상치 개수: {len(outliers)}")
    print(f"  이상치 비율: {ratio*100:.2f}%")
    print("-" * 50)
```

```python
# winsorizing 실시

from scipy.stats.mstats import winsorize

# winsorizing 적용할 비율 설정 (하위 5%, 상위 5%)
winsor_limits = (0.05, 0.05)

# 원본 데이터를 보존하기 위해 복사본 생성
winsorized_df = df.copy()

# 수치형 컬럼(정수형, 실수형) 선택
numeric_columns = winsorized_df.select_dtypes(include=['number']).columns

# 각 수치형 컬럼별 winsorizing 적용
for col in numeric_columns:
    # winsorize 함수는 masked array를 반환하므로 numpy 배열로 변환
    winsorized_values = winsorize(winsorized_df[col], limits=winsor_limits)
    winsorized_df[col] = np.array(winsorized_values)
```

```python
# winsorizing 후 이상치 판단

# int 타입 컬럼 선택 (int16, int32, int64 포함)
int_columns = winsorized_df.select_dtypes(include=['int64','float64']).columns

# 결과 저장을 위한 딕셔너리 (컬럼별 이상치 정보)
outlier_results = {}

for col in int_columns:
    series = df[col]
    # IQR 방식으로 경계 산출
    lb_iqr, ub_iqr = compute_iqr_bounds(series)
    
    # IQR 방식의 경계가 같으면, 분포가 매우 치우치거나 상수 데이터라고 판단하여 MAD 방식 적용
    if lb_iqr == ub_iqr:
        lb, ub = compute_mad_bounds(series)
        method = "MAD"
        outliers = series[(series < lb) | (series > ub)]
    else:
        lb, ub = lb_iqr, ub_iqr
        method = "IQR"
        outliers = series[(series < lb) | (series > ub)]
        
    ratio = len(outliers) / len(series)
    
    outlier_results[col] = {
        'method': method,
        'lower_bound': lb,
        'upper_bound': ub,
        'outliers': outliers.tolist(),
        'outlier_ratio': ratio
    }
    
    print(f"컬럼: {col} - 사용 방법: {method}")
    print(f"  하한: {lb}")
    print(f"  상한: {ub}")
    print(f"  이상치 개수: {len(outliers)}")
    print(f"  이상치 비율: {ratio*100:.2f}%")
    print("-" * 50)
    
# 일부는 이상치 비율을 줄였지만 그렇지 못한 경우가 많으므로 보류
```

```python
# Q1과 Q3가 같은 경우의 이상치는 중요한 지표로 쓰일 수 있어 값 그대로 보존
# Q1 Q3가 다른 컬럼에 한하여 전처리 진행
df_removed = df.copy()

# int 타입 컬럼 선택 (int16, int32, int64 포함)
int_columns = df_removed.select_dtypes(include=['int64','float64']).columns

# 각 int 컬럼별로 처리 수행
for col in int_columns:
    Q1 = df_removed[col].quantile(0.25)
    Q3 = df_removed[col].quantile(0.75)
    
    # IQR 방식이 적용 가능한 경우 : Q1과 Q3가 서로 다를 때
    if Q1 != Q3:
        # IQR 방식: 값이 Q3보다 큰 경우 Q3로, Q1보다 작은 경우 Q1으로 대체
        df_removed[col] = df_removed[col].clip(lower=Q1, upper=Q3)
        print(f"컬럼 '{col}': IQR 방식 적용 - 값이 {Q1:.2f} 미만은 {Q1:.2f}로, {Q3:.2f} 초과는 {Q3:.2f}로 대체")
    else:
        # MAD 전처리로 판단 (Q1과 Q3가 동일하므로 IQR 방식으로는 이상치 처리 불가)
        print(f"컬럼 '{col}': MAD 방식 적용 대상 - 값 변경 없이 그대로 유지")

# 처리 결과 확인 (예: 각 컬럼의 요약통계)
print("\n전처리 후 각 컬럼의 요약 통계량:")
print(df_removed[int_columns].describe())
```
