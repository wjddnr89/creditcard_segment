```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv(r'train_data(2)\train_회원정보.csv')
df_train
```

```python
cols = ['ID','Segment','연령','회원여부_이용가능','소지여부_신용',
        '소지카드수_유효_신용','회원여부_연체','유효카드수_신용',
        '이용가능카드수_신용','연회비발생카드수_B0M','카드신청건수','최종카드발급경과월']
```

```python
train_df = df_train[cols]
train_df
```

```python
train_df['연회비발생카드수_B0M'].value_counts()
```

```python
train_df['연령'].value_counts()
```

```python
train_df['연회비발생카드수_B0M']=train_df['연회비발생카드수_B0M'].map(
    {'0개' : 0, '1개이상' : 1}
)
```

```python
from sklearn.preprocessing import LabelEncoder
target_encoder = LabelEncoder()
train_df['Segment'] = target_encoder.fit_transform(train_df['Segment'])
```

```python
le1 = LabelEncoder()
train_df['연령'] = le1.fit_transform(train_df['연령'])
```

```python
from sklearn.ensemble import RandomForestClassifier
X_train, y_train = train_df.drop(columns=['Segment','ID'],axis=1),train_df['Segment']
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
test_df = pd.read_csv(r'C:\Users\user\OneDrive\Desktop\open\test_data(1)\test_회원정보.csv')
cols = ['ID','연령','회원여부_이용가능','소지여부_신용',
        '소지카드수_유효_신용','회원여부_연체','유효카드수_신용',
        '이용가능카드수_신용','연회비발생카드수_B0M','카드신청건수','최종카드발급경과월']
test_df= test_df[cols]
test_df['연회비발생카드수_B0M']=test_df['연회비발생카드수_B0M'].map(
    {'0개' : 0, '1개이상' : 1}
)
le2 = LabelEncoder()
test_df['연령'] = le2.fit_transform(test_df['연령'])
```

```python
y_pred = model.predict(test_df.drop('ID',axis=1))
y_pred
```

```python
import numpy as np

y_pred = np.array(pd.Series(y_pred).map({0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}))
y_pred
```

```python
# row 단위 예측 결과를 test_data에 추가
test_data = test_df.copy()  # 원본 유지
test_data["pred_label"] = y_pred
```

```python
test_data.columns
```

```python
submission = test_data.groupby("ID")["pred_label"] \
    .agg(lambda x: x.value_counts().idxmax()) \
    .reset_index()

submission.columns = ["ID", "Segment"]
submission.to_csv('./base_submit.csv',index=False)
```
