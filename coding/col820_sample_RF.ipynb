```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
import pandas as pd

# 1. 타겟 컬럼 설정
target = 'Segment'

# 2. 복사해서 원본 보호
df = train_seven.copy()

# 3. 수치형 피처 처리 (결측치 평균 대체)
df_numeric = df.select_dtypes(include=['number']).drop(columns=['ID'], errors='ignore')
imputer = SimpleImputer(strategy='mean')
df_numeric_imputed = pd.DataFrame(imputer.fit_transform(df_numeric), columns=df_numeric.columns)

# 4. 범주형 피처 인코딩
df_categoricals = df.select_dtypes(include=['object', 'category']).drop(columns=['ID'], errors='ignore')
for col in df_categoricals.columns:
    df_categoricals[col] = df_categoricals[col].astype('category').cat.codes

# 5. 합치기
df_all = pd.concat([df_numeric_imputed, df_categoricals], axis=1)

# 6. 타겟 분리
X = df_all.drop(columns=[target])
y = df_all[target]

# 7. 모델 학습
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
import platform
import matplotlib as mpl   
import matplotlib.pyplot as plt

# 운영체제에 맞게 한글을 지원하는 폰트를 설정합니다.
if platform.system() == 'Windows':
    # Windows의 경우 'Malgun Gothic'을 많이 사용합니다.
    mpl.rcParams['font.family'] = 'Malgun Gothic'


# 중요도 추출
importances = model.feature_importances_
feature_names = X.columns

# 중요도 상위 20개 시각화
importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
importance_df = importance_df.sort_values(by='importance', ascending=False).head(20)

# 시각화
plt.figure(figsize=(10, 6))
plt.barh(importance_df['feature'], importance_df['importance'])
plt.gca().invert_yaxis()
plt.xlabel("Importance")
plt.title("Top 20 Feature Importances from Random Forest")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# 1. 수치형 전처리 (결측치 평균 대체)
df_test_numeric = test_seven.select_dtypes(include=[np.number]).drop(columns=['ID'], errors='ignore')
df_test_numeric_imputed = pd.DataFrame(imputer.transform(df_test_numeric), columns=df_test_numeric.columns)

# 2. 범주형 인코딩
df_test_categoricals = test_seven.select_dtypes(include=['object', 'category']).drop(columns=['ID'], errors='ignore')
for col in df_test_categoricals.columns:
    df_test_categoricals[col] = df_test_categoricals[col].astype('category').cat.codes

# 3. 데이터 합치기
df_test_all = pd.concat([df_test_numeric_imputed, df_test_categoricals], axis=1)

# 4. 타겟 컬럼 제거 (혹시 있을 경우)
if 'Segment' in df_test_all.columns:
    df_test_all = df_test_all.drop(columns=['Segment'])

# 5. 예측 수행
y_pred = model.predict(df_test_all)

# 6. 예측 결과 숫자 → 라벨로 변환
label_map = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}
y_pred = np.array(pd.Series(y_pred).map(label_map))

# 7. 원본 test에 예측 추가
test_data = test_seven.copy()
test_data["pred_label"] = y_pred

# 8. ID별로 가장 많이 나온 label 선택 (ID가 중복된 경우만 필요)
submission = test_data.groupby("ID")["pred_label"] \
    .agg(lambda x: x.value_counts().idxmax()) \
    .reset_index()
submission.columns = ["ID", "Segment"]

# 9. 저장
submission.to_csv('./base_submit.csv', index=False)

# 10. 결과 확인
print("예측 결과 (y_pred):")
print(y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 예측값 (숫자 형태 그대로, 라벨 매핑 전)
y_pred_raw = model.predict(X)

# 실제값
y_true = y

# 1. accuracy
acc = accuracy_score(y_true, y_pred_raw)
print("✅ Accuracy:", round(acc, 4))

# 2. precision, recall, f1-score
print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred_raw, target_names=['A', 'B', 'C', 'D', 'E']))

# 3. confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred_raw)

plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
            xticklabels=['A', 'B', 'C', 'D', 'E'], 
            yticklabels=['A', 'B', 'C', 'D', 'E'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("🧩 Confusion Matrix")
plt.tight_layout()
plt.show()
```
